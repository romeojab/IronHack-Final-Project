```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get('https://play.google.com/store/apps/collection/cluster?clp=ogoqCBEqAggIMiIKHGNvbS5oYWxmYnJpY2suamV0cGFja2pveXJpZGUQARgD:S:ANO1ljKr-hA&gsr=Ci2iCioIESoCCAgyIgocY29tLmhhbGZicmljay5qZXRwYWNram95cmlkZRABGAM%3D:S:ANO1ljK4zAM')
wait_time = randint(3,6)
sleep(wait_time)
```

```python
%%time
# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")
 
while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
 
    # Wait to load page
    wait_time = randint(3,6)
    sleep(wait_time)
 
    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
driver
```

```python
%%time
links_games = []
elems = driver.find_elements(by=By.XPATH, value='//a[@href]') #find_elements_by_xpath is deprecated
for elem in elems:
    if 'details?id' in elem.get_attribute('href'):
        links_games.append((elem.get_attribute('href')))
        
links_games = list(dict.fromkeys(links_games))
```

```python
def clean_up(df):
    lst = ['in_app_purchases']
    for i in lst:
        df[[i, i + '_max']] = df[i].str.split('-', expand=True)
        df[i + '_max'] = df[i + '_max']
        
    cols = ['in_app_purchases','in_app_purchases_max']
    for col in cols:
        df[col] = df[col].str.strip('per item')
        df[col] = df[col].str.strip('€').astype('float')
    
    cols = ['installations']
    for col in cols:
        df[col] = df[col].str.strip('+')
        df[col] = df[col].str.replace(',','')
#         df[col] = df[col].str.strip(',')
    df = df.rename(columns={'in_app_purchases': 'in_app_purchases_min'})
    df['in_app_purchases_max'] = np.where(df['in_app_purchases_max'] == 0,
                                          df['in_app_purchases_min'],df['in_app_purchases_max']) 
    return df                            
```
